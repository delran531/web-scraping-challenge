````python
# Dependencies
import pandas as pd
import requests
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup as bs
````

````python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
print(soup.body.prettify())
````

````python
# Find articles
articles = soup.find_all('div', class_='content_title')
articles
````

````python
# Find paragraph text
article_p = soup.find_all('div', class_='rollover_description_inner')
article_p
````

````python
# Scrape the [NASA Mars News Site](https://mars.nasa.gov/news/) and 
# collect the latest News Title and Paragraph Text. Assign the text 
# to variables that you can reference later
news_title = articles[0].text
news_p = article_p[0].text

print(news_title)
print(news_p)
````

````python
### Mars Hemispheres

# * Visit the USGS Astrogeology site [here]
# (https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars) 
# to obtain high resolution images for each of Mar's hemispheres.

# * You will need to click each of the links to the hemispheres in order to find the 
# image url to the full resolution image.

# * Save both the image url string for the full resolution hemisphere image, and the 
# Hemisphere title containing the hemisphere name. Use a Python dictionary to store the data 
# using the keys `img_url` and `title`.

# * Append the dictionary with the image url string and the hemisphere title to a list. 
# This list will contain one dictionary for each hemisphere.

# ```python
# # Example:
# hemisphere_image_urls = [
#     {"title": "Valles Marineris Hemisphere", "img_url": "..."},
#     {"title": "Cerberus Hemisphere", "img_url": "..."},
#     {"title": "Schiaparelli Hemisphere", "img_url": "..."},
#     {"title": "Syrtis Major Hemisphere", "img_url": "..."},
# ]
# ```
````

````python
# Use splinter to get image and title links of each hemisphere
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
 
usgs_page = Browser('chrome', headless=False)
usgs_page.visit(url)
````

````python
# Parse html file with BeautifulSoup
mars_hemis_html = usgs_page.html
mars_hemis_soup = bs(mars_hemis_html, 'html.parser')
````

````python

print(mars_hemis_soup.body.prettify())
````

````python
# Find pic
mars_hemis = mars_hemis_soup.find_all('div', class_='description')
mars_hemis
````

````python
mars_hemis[0]
````

````python
mars_hemis[0].find('a', class_='itemLink')

````

````python
hemi_image_urls = []

for image in mars_hemis:
    hemi_url = image.find('a', class_='itemLink')
    hemi = hemi_url.get('href')
    hemi_link = 'https://astrogeology.usgs.gov' + hemi
    print(hemi_link)

    # Visit each link that you just found (hemisphere_link)
    usgs_page.visit(hemi_link)
    
    # Create dictionary to hold title and image url
    hemi_image_dict = {}
    
    # Need to parse html again
    mars_hemis_html = usgs_page.html
    mars_hemis_soup = bs(mars_hemis_html, 'html.parser')
    
    # Get image link
    hemi_link = mars_hemis_soup.find('a', text='Original').get('href')
    
    # Get title text
    hemi_title = mars_hemis_soup.find('h2', class_='title').text.replace(' Enhanced', '')
    
    # Append title and image urls of hemisphere to dictionary
    hemi_image_dict['title'] = hemi_title
    hemi_image_dict['img_url'] = hemi_link
    
    # Append dictionaries to list
    hemi_image_urls.append(hemi_image_dict)

print(hemi_image_urls)
````
